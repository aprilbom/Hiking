```python
from PIL import Image, ImageOps
import os, sys
import glob
```

```python
MtList = ['InWang', 'GwanAk', 'DoBong', 'BukAk', 'CheongGye']
MtFolderList = ['InWang', 'GwanAk', 'DoBong', 'BukAk', 'CheongGye']

for MtFolder in MtFolderList:
    image_dir = "./data/" + MtFolder + "/"
    target_resize_dir = "./data_resize/" + MtFolder + "/"
    target_rotate_dir = "./data_rotate/" + MtFolder + "/"
    target_invert_dir = "./data_invert/" + MtFolder + "/"
    target_flip_dir = "./data_flip/" + MtFolder + "/"
    
    # 이미지 리사이징
    if not os.path.isdir(target_resize_dir):
        os.makedirs(target_resize_dir)
    
    if not os.path.isdir(target_rotate_dir):
        os.makedirs(target_rotate_dir)
        
    if not os.path.isdir(target_invert_dir):
        os.makedirs(target_invert_dir)
        
    if not os.path.isdir(target_flip_dir):
        os.makedirs(target_flip_dir)
    
    files = glob.glob(image_dir + "*.*")
    
    # 폴더 내의 이미지 갯수 확인
    print(len(files))
    
    count = 1
    
    size = (244, 244)
    
    # 각 폴더마다 진행
    for file in files:
        im = Image.open(file)
        im = im.convert('RGB')

        print("i: ", count, im.format, im.size, im.mode, file.split("ㅊ")[-1])

        count += 1

        # 이미지 자르기 (244, 244)
        im = ImageOps.fit(im, size, Image.ANTIALIAS, 0, (0.5, 0.5))
        im.save(target_resize_dir + file.split("\\")[-1].split(".")[0] + ".jpg", quality = 100)
        
        # 이미지 흑백처리
        invert_im = ImageOps.invert(im)
        invert_im.save(target_invert_dir + file.split("\\")[-1].split(".")[0] + ".jpg", quality = 100)
        
        # 이미지 돌리기
        im.rotate(90).save(target_rotate_dir + "resize_" + file.split("\\")[-1].split(".")[0] + ".jpg", quality = 100)
        
        # 이미지 플립
        flip_im = ImageOps.mirror(im)
        flip_im.save(target_flip_dir + file.split("\\")[-1].split(".")[0] + ".jpg", quality = 100)
```
